```python
import pandas as pd
import numpy as np
# from google.colab import files
# files.upload()
```

```python
data = pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
data.info()
```

```python
df = data.iloc[ : , 3 : ]
df.head()
```

```python
df['Exited'].value_counts()
```

```python
df
```

```python
df = pd.get_dummies(df, columns = ['Geography' , 'Gender'], drop_first = True , dtype = int)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns = ['Exited'] , axis = 0),df['Exited'] , train_size= 0.8 , random_state= True)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scalar = scaler.fit_transform(x_train)
x_test_scalar = scaler.transform(x_test)
```

```python
x_train_scalar
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(3,activation = 'sigmoid' , input_dim = 11))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy' , optimizer = 'Adam')
```

```python
history= model.fit(x_train_scalar,y_train,epochs = 10)
```

```python
model.layers[0].get_weights()
```

```python
model.layers[1].get_weights()
```

```python
y = model.predict(x_test_scalar)
```

```python
y_test.value_counts()
```

```python
y_pred = np.where(y > 0.5,1,0)
```

```python
y_pred
```

```python
np.sum(y_pred == 1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
